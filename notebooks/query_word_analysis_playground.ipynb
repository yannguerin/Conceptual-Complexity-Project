```python
in_definition_value_counts = df.word_in_definition.value_counts()
```

```python
in_definition_value_counts.head(50)
```

```python
in_definition_value_counts.tail(50)
```

```python
in_definition_counts_df = in_definition_value_counts.rename_axis('word').to_frame('frequency')
```

```python
in_definition_counts_df
```

```python
def index_value(r):
    return 1 / r
```

```python
in_definition_value_index = pd.DataFrame(in_definition_counts_df["frequency"].map(index_value), columns=['frequency'])
```

```python
in_definition_value_index.head(50)
```

```python
in_definition_value_index.columns
```

```python
muon_exp = "intensity variation of high-energy positron from muon decay directly encode the difference frequency"
science_exp = "science is a systematic endeavor that build and organize knowledge in the form of test explanation and prediction about the universe"
small_talk = "hello friend how are you today i am doing very well and i am happy"
fake_complex = "apostrophe complexity cataclysm dire defenestration archaic armchair"
```

```python
index = 0
count = 0
for word in muon_exp.split():
    try:
        index += float(in_definition_value_index.frequency[in_definition_value_index.index == word])
        count += 1
    except TypeError:
        print(word)
        
print(index / count)
```

```python
def complexity_index(text: str):
    index = 0
    count = 0
    for word in text.split():
        try:
            index += float(in_definition_value_index.frequency[in_definition_value_index.index == word])
            count += 1
        except TypeError:
            print(word)
    
    return index / count
```

```python
print(complexity_index(science_exp))
```

```python
print(complexity_index(muon_exp))
```

```python
print(complexity_index(small_talk))
```

```python
print(complexity_index(fake_complex))
```
